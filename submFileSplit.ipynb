```python
import pandas as pd

df_subm1 = pd.read_csv('subm1_labels_revealed.csv', sep=';')
df_subm2 = pd.read_csv('subm2_labels_revealed.csv', sep=';')


df_final = pd.concat([df_subm1, df_subm2]).drop_duplicates(subset=["Text"]).reset_index(drop=True)
df_final.shape

df_texts = df_final[["ID","Text"]]
df_labels = df_final["Label"]

df_texts.to_csv('everySubm_revealed_texts.csv', index=False)
df_labels.to_csv('everySubm_revealed_labels.csv', index=False)



```
